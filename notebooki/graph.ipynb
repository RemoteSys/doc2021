```python
import graphviz
from graphviz import Digraph
```

```python
gr = Digraph()
type(gr)
```

```python
idx = [f'n{i}' for i in range(1,6)]
naz = [f'node_{i}' for i in range(1,6)]
idx, naz
```

```python
for i,n in zip(idx,naz):
    gr.node(i,n)
```

```python
ed = [('n1','n3'),('n2','n4'),('n2','n5')]
gr.edges(ed)
gr
```

```python
print(gr.source)
```

```python
gr.node('n1',shape='doubleoctagon',color='red')
gr
```

```python
gr.edge('n1','n2', arrowsize='2', arrowhead ='crow')
gr
```

```python
gr.render('blok1',format='pdf')
gr.render('blok1',format='png')
```
