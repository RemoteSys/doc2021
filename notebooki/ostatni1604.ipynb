```python
im2 = Image.open('../data/ob1.png')
print(type(im2))
plt.imshow(im2);
```

### Podstawowe przekształcenia obrazu

```python
def rot(a,ex):
    new = im2.rotate(a,expand=ex)
    plt.imshow(new)
    
interact(rot, a=(0,180,15),ex=[True,False])
```

```python
# zmiana rozmiaru obrazu (zmiana rozdzielczości)
def res(a,b):
    new = im2.resize((a,b))
    f,ax = plt.subplots(1,2,figsize=(12,6))
    ax[0].imshow(im2)
    ax[1].imshow(new)
    
interact(res, a=(100,400,50),b=(100,400,50))
```

```python
# wycięcie fragmentu obrazu
bb = (150,150,400,300)
new = im2.crop(bb)
plt.imshow(new);
```

# 4. scikit

### k-maens

```python
from sklearn.cluster import KMeans
```

```python
dane = ar.reshape(-1,3).copy()
ar.shape, dane.shape
```

```python
km = KMeans(n_clusters=4, random_state=0).fit(dane)
label = km.labels_
label.shape
```

```python
new = label.reshape(ar.shape[0],ar.shape[1])

f,ax = plt.subplots(2,2,figsize=(8,8))
ax[0][0].imshow(ar)
ax[0][1].hist(ar.ravel(),bins=256)
ax[1][0].imshow(new,cmap=plt.cm.Greys_r)
ax[1][1].hist(new.ravel(),bins=10)
```

```python
# zobacz tylko wybranę klasę np. 2
tmp = new.copy()
idx = tmp != 2
tmp[idx] = 0
plt.imshow(tmp,cmap=plt.cm.Greys_r)
```
